```python
# This is where the practice pipeline starts

"""
1. Mitochondrial annotation (important fix)
2. QC metrics
3. Inspect QC structure
4. QC visualization (critical checkpoint)
5. Scatter QC relationship
6. Filtering (core preprocessing decision)
7. Normalize + log transform
8. Highly variable genes (very important)
9. Subset to HVGs
10. PCA
11. Neighborhood graph + UMAP
12. Clustering
13. Differential expression (biological signal)
"""
```

```python
# Step 1 — Mitochondrial annotation

import scanpy as sc

adata.var["mt"] = adata.var_names.str.startswith("MT-")
```

```python
# Step 2 — QC metrics

sc.pp.calculate_qc_metrics(
    adata,
    qc_vars=["mt"],
    inplace=True
)
```

```python
# Step 3 — Inspect QC structure

adata.obs.head()
```

```python
# Step 4 — QC visualization (critical checkpoint)

sc.pl.violin(
    adata,
    [
        "n_genes_by_counts",
        "total_counts",
        "pct_counts_mt"
    ],
    jitter=0.4,
    multi_panel=True
)
```

```python
# Step 5 — Scatter QC relationship

sc.pl.scatter(
    adata,
    x="total_counts",
    y="n_genes_by_counts",
    color="pct_counts_mt"
)
```

```python
# Step 6 — Filtering (core preprocessing decision)

adata = adata[
    adata.obs.n_genes_by_counts > 200,
    :
]

adata = adata[
    adata.obs.n_genes_by_counts < 6000,
    :
]

adata = adata[
    adata.obs.pct_counts_mt < 10,
    :
]
```

```python
# Step 7 — Normalize + log transform

sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)
```

```python
# This is where the return to the pipeline begins
```

```python
# Step 8 - Highly variable genes

sc.pp.highly_variable_genes(
    adata,
    n_top_genes=2000,
    batch_key="sample"
)
```

```python
# Step 9 — Subset to HVGs

adata = adata[:, adata.var.highly_variable]
```

```python
# Step 10 — PCA

sc.pp.scale(adata, max_value=10)

sc.tl.pca(adata)
sc.pl.pca_variance_ratio(adata)
```

```python
# Step 11 — Neighborhood graph + UMAP

sc.pp.neighbors(adata, n_neighbors=15, n_pcs=30)

sc.tl.umap(adata)

sc.pl.umap(adata, color=["condition", "sample"])
```

```python
# Step 12 — Clustering

sc.tl.leiden(adata, resolution=0.5)

sc.pl.umap(adata, color="leiden")
```

```python
# Step 13 — Differential expression (biological signal)

sc.tl.rank_genes_groups(
    adata,
    groupby="condition",
    method="wilcoxon"
)

sc.pl.rank_genes_groups(adata, n_genes=25)
```
